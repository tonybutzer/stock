```python
# read csv file
motley = pd.read_csv("/data/tonyindex.csv")

# display DataFrame
motley
```

```python
cdf  = pd.DataFrame()
for sym in motley['Symbol']:
    print(sym)
    mdf = cars.loc[cars['Symbol'] == sym]
    print(mdf)
    cdf = cdf.append(mdf)
```
